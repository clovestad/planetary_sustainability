define planet sustainibility criteria
 
    Distance: Below 1000 light-years.
    Atmospheric Oxygen Content: Above 0.5.
    Soil Fertility: Above 0.5.
    Median Seasonal Temperatures: Between 0°F and 100°F.
    Availability of Potable Water: Yes.
    Suitability for Life: No extreme conditions.


```python
# Adjust filtering criteria based on the actual column names
criteria = (
    (df["Distance (light-years)"] < 1000)
    & (df["Atmospheric Oxygen Content"] > 0.5)
    & (df["Soil Fertility"] > 0.5)
    & (0 <= df["Median Seasonal Temperatures (F)"]) & (df["Median Seasonal Temperatures (F)"] <= 100)
    & (df["Availability of Potable Water"] == "Y")
    & (df["Suitability for Life"] == "No")
)

# Apply the filter to create a new DataFrame
livable_planets_df = df[criteria].copy()


```
